```python
import cv2
from deepface import DeepFace

import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
backends = ['opencv', 'ssd', 'dlib', 'mtcnn']
```

```python
faces = DeepFace.extract_faces("D:/Hackathon/Image Classification/Images data 1/Image classification/Dataset/me/IMG_1380.JPG", detector_backend='opencv')
```

```python
faces
```

```python
if faces:
    face_image = faces[0]['face']

    plt.imshow(face_image)
    plt.axis('off') 
    plt.show()
else:
    print("No faces detected.")
```

```python
fig, axs = plt.subplots(2, 2, figsize = (15,10))

axs = axs.flatten()

for i,b in enumerate(backends):
    try:
        axs[i].set_title(b)
        face = DeepFace.extract_faces("D:/Hackathon/Image Classification/Images data 1/Image classification/Dataset/me/IMG_1380.JPG", detector_backend=b)
        axs[i].imshow(face[0]['face'])
        axs[i].axis('off')
    except:
        pass

plt.show()
```

```python
models = ["VGG-Face", "Facenet", "Facenet512", "OpenFace"]
for model in models:
    result = DeepFace.verify("D:/Hackathon/Image Classification/Images data 1/Image classification/Dataset/me/IMG_1380.JPG",
                         "D:\Hackathon/Image Classification/Images data 1/Image classification/Dataset/me/IMG_7244-1.jpg",
                         model_name = model)
    fig, axs = plt.subplots(1,2, figsize=(10,5))
    axs[0].imshow(plt.imread("D:/Hackathon/Image Classification/Images data 1/Image classification/Dataset/me/IMG_1380.JPG"))
    axs[1].imshow(plt.imread("D:\Hackathon/Image Classification/Images data 1/Image classification/Dataset/me/IMG_7244-1.jpg"))
    fig.suptitle(f'Verified {result['verified']} - Distance {result['distance']:0.4} Model = {model}')

    axs[0].axis('off')
    axs[1].axis('off')

plt.show()
```
